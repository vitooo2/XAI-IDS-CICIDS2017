```python
import pandas as pd
import glob
import numpy as np

files = glob.glob("*.csv")

df_list = []

for file in files:

    temp_df = pd.read_csv(file)

    df_list.append(temp_df)

df = pd.concat(
    df_list,
    ignore_index=True
)

print("\nDataset Shape:")
print(df.shape)
```

```python
print(
    df.isnull().sum().sum()
)
```

```python
print(
    df[" Label"].value_counts()
)
```

```python
missing_values = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing_values[
    missing_values > 0
]

df = df.dropna()

print(df.shape)

# Remove Infinity Values


df.replace([np.inf, -np.inf], np.nan, inplace=True)

df.dropna(inplace=True)

print("Dataset Shape After Cleaning:", df.shape)
```

```python
print(
    df[" Label"].value_counts()
)
```

```python
df = df[
    ~df[" Label"].isin([
        "Heartbleed",
        "Infiltration"
    ])
]
df[" Label"].value_counts()
```

```python
print(df.shape)

print(df[" Label"].value_counts())
```

```python
attack_mapping = {

    "BENIGN": "Benign",

    # DoS
    "DDoS": "DoS",
    "DoS Hulk": "DoS",
    "DoS GoldenEye": "DoS",
    "DoS slowloris": "DoS",
    "DoS Slowhttptest": "DoS",

    # Brute Force
    "FTP-Patator": "BruteForce",
    "SSH-Patator": "BruteForce",

    # Port Scan
    "PortScan": "PortScan",

    # Bot
    "Bot": "Bot",

    # Web Attack
    "Web Attack � Brute Force": "WebAttack",
    "Web Attack � XSS": "WebAttack",
    "Web Attack � Sql Injection": "WebAttack"
}

df[" Label"] = df[" Label"].map(attack_mapping)
```

```python
print(df[" Label"].value_counts())
```

```python
balanced_df = (
    df.groupby(" Label", group_keys=False)
      .apply(
          lambda x: x.sample(
              n=1900,
              random_state=42
          )
      )
      .reset_index(drop=True)
)
```

```python
print(
    balanced_df[" Label"].value_counts()
)
```

```python
print(balanced_df.shape)
```

```python
balanced_df.to_csv(
    "CIC_IDS2017_balanced2.csv",
    index=False
)
```
